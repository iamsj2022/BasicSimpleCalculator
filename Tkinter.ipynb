```python
#Using get function


from tkinter import *

root = Tk()

e = Entry(root,width = 50 ,borderwidth = 5)
e.pack()
e.insert(0,"Enter Your Name :  ")

def myClick():
    myLabel = Label(root,text = e.get())
    myLabel.pack()
    

myButton = Button(root,text = "Enter Your Name",padx = 50 ,pady = 10,command = myClick,fg = "Red",bg = "Black")

myButton.pack()



root.mainloop()

```

```python

from tkinter import *

root = Tk()
root.title('Simple Calculator')
root.iconbitmap('C:\\Shashi Jha DS--ML--DL__NLP__CV\\SJ.PNG')





e = Entry(root, width = 30 ,borderwidth = 5)
e.grid(row = 0,column = 0,columnspan = 3,padx = 10,pady =10)




def Button_click(number):
    current = e.get()
    e.delete(0, END)
    e.insert(0,str(current) + str(number))

def Button_clear() :
     e.delete(0, END)
        
def Button_Add():
    first_number = e.get()
    global f_num
    global math 
    math = "Addition"
    f_num = int(first_number)
    e.delete(0, END)

def Button_Sub():
    first_number = e.get()
    global f_num
    global math 
    math = "Subtraction"
    f_num = int(first_number)
    e.delete(0, END)
    
def Button_Mul():
    first_number = e.get()
    global f_num
    global math 
    math = "Multiplication"
    f_num = int(first_number)
    e.delete(0, END)    

def Button_Div():
    first_number = e.get()
    global f_num
    global math 
    math = "Division"
    f_num = int(first_number)
    e.delete(0, END)
    
def Button_Equal() :
    second_number = e.get()
    e.delete(0, END)
    
    if math == "Addition" :
        e.insert(0,f_num + int(second_number))
        
    if math == "Subtraction" :
        e.insert(0,f_num - int(second_number))

    if math == "Multiplication" :
        e.insert(0,f_num * int(second_number))
        
    if math == "Division" :
        e.insert(0,f_num / int(second_number))
    
    
# Defining Buttons
myButton_1 = Button(root,text = "1",padx = 50 ,pady = 10,command =lambda: Button_click(1),fg = "Red",bg = "Black").grid(row=3,column=0)
myButton_2 = Button(root,text = "2",padx = 50 ,pady = 10,command =lambda: Button_click(2),fg = "Red",bg = "Black").grid(row=3,column=1)
myButton_3 = Button(root,text = "3",padx = 50 ,pady = 10,command =lambda: Button_click(3),fg = "Red",bg = "Black").grid(row=3,column=2)

myButton_4 = Button(root,text = "4",padx = 50 ,pady = 10,command =lambda: Button_click(4),fg = "Red",bg = "Black").grid(row=2,column=0)
myButton_5 = Button(root,text = "5",padx = 50 ,pady = 10,command =lambda: Button_click(5),fg = "Red",bg = "Black").grid(row=2,column=1)
myButton_6 = Button(root,text = "6",padx = 50 ,pady = 10,command =lambda: Button_click(6),fg = "Red",bg = "Black").grid(row=2,column=2)

myButton_7 = Button(root,text = "7",padx = 50 ,pady = 10,command =lambda: Button_click(7),fg = "Red",bg = "Black").grid(row=1,column=0)
myButton_8 = Button(root,text = "8",padx = 50 ,pady = 10,command =lambda: Button_click(8),fg = "Red",bg = "Black").grid(row=1,column=1)
myButton_9 = Button(root,text = "9",padx = 50 ,pady = 10,command =lambda: Button_click(9),fg = "Red",bg = "Black").grid(row=1,column=2)

myButton_0 = Button(root,text = "0",padx = 50 ,pady = 10,command =lambda: Button_click(0),fg = "Red",bg = "Black").grid(row=4,column=0)
myButton_Add = Button(root,text = "+",padx = 50 ,pady = 10,command =Button_Add,fg = "Red",bg = "Black").grid(row=4,column=1)
myButton_Sub = Button(root,text = "-",padx = 50 ,pady = 10,command =Button_Sub,fg = "Red",bg = "Black").grid(row=4,column=2)

myButton_Mul = Button(root,text = "*",padx = 50,pady = 10,command =Button_Mul,fg = "Red",bg = "Black").grid(row=5,column=0)
myButton_Div = Button(root,text = "/",padx = 50 ,pady = 10,command =Button_Div,fg = "Red",bg = "Black").grid(row=5,column=1)
button_exit = Button(root,text = "EXIT",padx = 50 ,pady = 10,command =root.quit,fg = "Red",bg = "Black").grid(row=5,column=2)

myButton_Clear = Button(root,text = "Clear",padx = 45 ,pady = 10,command =Button_clear,fg = "Red",bg = "Black").grid(row=6,column=0)
myButton_Equal = Button(root,text = "=",padx = 50 ,pady = 10,command =Button_Equal,fg = "Red",bg = "Black").grid(row=6,column=1)
myButton_SJ = Button(root,text = "SJ",padx = 50 ,pady = 10,command =Button_click,fg = "Red",bg = "Black").grid(row=6,column=2)







root.mainloop()

```
